```python
#extrae distancias entre cada ciudad
def generar_coordenadas_no_euclidianas(df, n=20):
    required_columns = {'city', 'lat', 'lng'}
    if not required_columns.issubset(df.columns):
        raise ValueError(f"El DataFrame debe contener las columnas: {required_columns}")

    ciudades = df.head(n).reset_index(drop=True)
    nombres = ciudades['city'].tolist()
    num_ciudades = len(ciudades)
    distancias = np.zeros((num_ciudades, num_ciudades))

    for i in range(num_ciudades):
        coord_i = (ciudades.loc[i, 'lat'], ciudades.loc[i, 'lng'])
        for j in range(i + 1, num_ciudades):
            coord_j = (ciudades.loc[j, 'lat'], ciudades.loc[j, 'lng'])
            distancia = geodesic(coord_i, coord_j).kilometers
            distancias[i][j] = distancia
            distancias[j][i] = distancia

    return np.round(distancias, 2), nombres

def generar_coordenadas(df, n=25):
    required_columns = {'city', 'lat', 'lng'}
    if not required_columns.issubset(df.columns):
        raise ValueError(f"El DataFrame debe contener las columnas: {required_columns}")
    ciudades = df.head(n).reset_index(drop=True)
    nombres = ciudades['city'].tolist()
    num_ciudades = len(ciudades)
    distancias = np.zeros((num_ciudades, num_ciudades))
    for i in range(num_ciudades):
        coord_i = (ciudades.loc[i, 'lat'], ciudades.loc[i, 'lng'])
        for j in range(i + 1, num_ciudades):
            coord_j = (ciudades.loc[j, 'lat'], ciudades.loc[j, 'lng'])
            distancia = np.sqrt((coord_j[0] - coord_i[0])**2 + (coord_j[1] - coord_i[1])**2)
            distancias[i][j] = distancia
            distancias[j][i] = distancia
    return np.round(distancias, 2), nombres
```

```python
#una hormiga se mueve de i a j
def pick_move(current, visited, distancias, pheromone, alpha, beta, save_probs=False):
    probabilities = []
    nodes = []
    total = 0
    prob_dict = {}  # Para guardar la probabilidad asociada a cada nodo

    # Calcular tau e eta y acumular las probabilidades
    for node in range(len(distancias)):
        if node not in visited:
            tau = pheromone[current][node] ** alpha
            eta = (1.0 / distancias[current][node]) ** beta
            prob = tau * eta
            probabilities.append(prob)
            nodes.append(node)
            total += prob

    if total == 0:
        # Todas las probabilidades son cero: elegir al azar entre los no visitados
        chosen = random.choice([node for node in range(len(distancias)) if node not in visited])
        if save_probs:
            return chosen, {}
        return chosen

    # Normalizar las probabilidades
    probabilities = [p / total for p in probabilities]

    # Guardar las probabilidades si se solicita
    if save_probs:
        for i, node in enumerate(nodes):
            prob_dict[node] = probabilities[i]

    # Elegir el siguiente nodo basado en las probabilidades
    chosen = random.choices(nodes, weights=probabilities)[0]

    if save_probs:
        return chosen, prob_dict
    return chosen
```

```python
#una hormiga se mueve por todas las ciudades hasta completar el recorrido y cerrarlo
def gen_path(start, distancias, pheromone, alpha, beta):
    path = [start]
    visited = set([start])
    current_city = start

    prob_history = []  # Lista para guardar las probabilidades paso a paso

    for _ in range(len(distancias) - 1):
        next_city, probs = pick_move(current_city, visited, distancias, pheromone, alpha, beta, save_probs=True)

        # Guardar el paso y sus probabilidades
        prob_history.append({
            "from": current_city,
            "to": next_city,
            "probabilities": probs
        })

        path.append(next_city)
        visited.add(next_city)
        current_city = next_city

    path.append(start)  # Regresamos al inicio
    return path, prob_history
```

```python
#obtiene el camino para las n_hormigas
def caminos(n_ants, distances, pheromone, alpha, beta):
    results = []
    for _ in range(n_ants):
        path, prob_history = gen_path(0, distancias, pheromone, alpha, beta)
        results.append({
            "path": path,
            "prob_history": prob_history
        })

    return results
```

```python
#obtiene el costo y el trazo de los caminos para todas las hormigas por cada ciclo
def generar_paths_y_costos(n_ants, distancias, pheromone, alpha, beta):
    resultados = []

    for _ in range(n_ants):
        path, prob_history = gen_path(0, distancias, pheromone, alpha, beta)

        # Calcular costo total del tour
        costo = 0
        for i in range(len(path) - 1):
            costo += distancias[path[i]][path[i+1]]

        # Guardar toda la info de la hormiga
        resultados.append({
            "path": path,
            "cost": costo,
            "prob_history": prob_history
        })

    return resultados
```

```python
#actualiza la feromona despues de cada ciclo
n_iterations = 100
n_ants = 100
alpha = 1
beta = 0.8
Q = 10
decay = 0.5
pheromone = np.ones((len(distancias), len(distancias)))

def update_feromonas(pheromone, hormigas, Q, decay):
    pheromone *= (1 - decay)  # Evaporación

    for hormiga in hormigas:
        camino = hormiga["path"]
        costo = hormiga["cost"]
        delta = Q / costo

        for i in range(len(camino) - 1):
            a, b = camino[i], camino[i + 1]
            pheromone[a][b] += delta
            pheromone[b][a] += delta  # Simétrico

    return pheromone


feromonas_historial = []
for iter in range(n_iterations):
    hormigas = generar_paths_y_costos(n_ants, distancias, pheromone, alpha, beta)
    pheromone = update_feromonas(pheromone, hormigas, Q, decay)
    feromonas_historial.append(pheromone.copy())  # Guardar copia de la matriz

```

```python
#ejecuta todos los ciclos
def run_ACO(n_iterations, n_ants, distances, pheromone, alpha, beta, Q, decay, verbose=True, guardar_historia=True):
    mejor_costo = float("inf")
    mejor_camino = None

    peor_costo = float("-inf")  # 👈 Para peor costo global
    peor_camino = None          # 👈 Para peor camino global

    historial_costos = []          # Mejor costo por iteración
    historial_peores_costos = []   # Peor costo por iteración
    historial_feromonas = []

    for i in range(n_iterations):
        # Generar caminos y calcular costos
        hormigas = generar_paths_y_costos(n_ants, distances, pheromone, alpha, beta)

        # Actualizar feromonas
        pheromone = update_feromonas(pheromone, hormigas, Q, decay)

        # Guardar feromonas
        if guardar_historia:
            historial_feromonas.append(pheromone.copy())

        # Evaluar costos
        costos = [h["cost"] for h in hormigas]
        caminos = [h["path"] for h in hormigas]
        costo_min = min(costos)
        camino_min = caminos[costos.index(costo_min)]

        costo_max = max(costos)  # 👈 Peor costo de la iteración
        camino_max = caminos[costos.index(costo_max)]

        # Guardar mejores y peores costos por iteración
        historial_costos.append(costo_min)
        historial_peores_costos.append(costo_max)

        # Actualizar mejor camino global
        if costo_min < mejor_costo:
            mejor_costo = costo_min
            mejor_camino = camino_min

        # Actualizar peor camino global
        if costo_max > peor_costo:
            peor_costo = costo_max
            peor_camino = camino_max

        if verbose:
            print(f"Iteración {i+1}/{n_iterations} - Mejor costo: {costo_min} - Peor costo: {costo_max}")

    return {
        "pheromone": pheromone,
        "mejor_costo": mejor_costo,
        "mejor_camino": mejor_camino,
        "peor_costo": peor_costo,       # 👈 Devuelvo peor costo global
        "peor_camino": peor_camino,     # 👈 Devuelvo peor camino global
        "historial_costos": historial_costos,
        "historial_peores_costos": historial_peores_costos,
        "historial_feromonas": historial_feromonas if guardar_historia else None
    }

```

```python
# Ejecutar el ACO
#Hiperparametros
import time
inicio = time.time()

feromona_inicial=0.1 # feromona de partida
n_ciudades = 25
pheromone = np.ones((n_ciudades, n_ciudades)) * feromona_inicial

resultados = run_ACO(
    n_iterations=100,
    n_ants=300,
    distances=distancias,
    pheromone=pheromone,
    alpha=1,
    beta=0.8,
    Q=10,
    decay=0.1
)

print("\n Mejor camino encontrado:")
print(resultados["mejor_camino"])
print(f"Costo total: {resultados['mejor_costo']}")

print("\n Peor camino encontrado:")
print(resultados["peor_camino"])
print(f"Costo total: {resultados['peor_costo']}")

fin = time.time()
print(f"Tiempo de ejecución: {fin - inicio:.4f} segundos")
```

```python
def mostrar_ruta_folium(df, nombres, mejor_ruta, peor_ruta, pausa_segundos=1):
    # Centrar el mapa en el inicio de la mejor ruta
    lat_centro = df.loc[mejor_ruta[0], 'lat']
    lng_centro = df.loc[mejor_ruta[0], 'lng']
    mapa = folium.Map(location=[lat_centro, lng_centro], zoom_start=6)    # Coordenadas de las rutas
    coords_mejor = [(df.loc[i, 'lat'], df.loc[i, 'lng']) for i in mejor_ruta]
    coords_peor = [(df.loc[i, 'lat'], df.loc[i, 'lng']) for i in peor_ruta]    # ===== Marcadores para peor ruta (rojos) =====
    for idx in peor_ruta:
        folium.Marker(
            location=[df.loc[idx, 'lat'], df.loc[idx, 'lng']],
            popup=f"Peor: {nombres[idx]}",
            icon=folium.Icon(color='red', icon='info-sign')
        ).add_to(mapa)    # ===== Animación peor ruta =====
    for i in range(len(coords_peor) - 1):
        segmento = coords_peor[i:i+2]
        AntPath(
            locations=segmento,
            color='red',
            weight=5,
            delay=1000
        ).add_to(mapa)
        display(mapa)
        time.sleep(pausa_segundos)
        clear_output(wait=True)
```
